```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Preprocessing

```python
df = pd.read_csv("Credit_Card.csv")
display(df)
df = df.dropna()
df = df.drop(columns=['CUST_ID'])  #Drop Column CUST_ID
df = df[df['TENURE'] >= 10]  #Drop rows with TENURE less than 10
df
```

# TENURE as target variable, Rest of the columns as Features

```python
X = df.drop(columns = ['TENURE'])
y = df["TENURE"]
```

# Normalize the features

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
print(X_scaled)
print(X_scaled.shape)
print("Dimensions of feature set = ",X_scaled.shape[1])
```

# The Dimensions of the Feature Set = 16

# Principal Component Analysis (Converting to 2-Dimensional Feature set)

```python
cm = np.cov(X_scaled.T)
display(cm)
display(cm.shape)
```

```python
eig_val, eig_vec = np.linalg.eig(cm)
display(eig_val)
display(eig_vec)
```

```python
tot = sum(eig_val)
display(tot)
```

```python
sorted_eig_val = [i for i in sorted(eig_val, reverse = True)]
display(sorted_eig_val)
```

```python
# Explained Varinace
exp_var = [(i / tot) for i in sorted_eig_val]
display(exp_var)
cum_exp_var = np.cumsum(exp_var)
display(cum_exp_var)
plt.bar(range(1, 17), exp_var, label = "Explained Variance")
plt.xlabel("Principal Components")
plt.ylabel("Explained Variance");
```

## The transformed Features of Dim=2 using PCA help us visualize the higer dimension data better. the PCA1 (x-axis) shows explains the most variance in the data hence the points are scatter most along the x direction. PCA 2 (y-axis) explains the 2nd most variance. Together the variables explain more than 50% of the variance in the data.
## The model is not able to segregate the data into clusters in any meaningful way. 


```python
# Choosing Dimensions = 2, Taking the pair of Eigen Values and Eigen Vectors

eigen_pair = [(np.abs(eig_val[i]), eig_vec[:, i]) for i in range(len(eig_val))]
display(eigen_pair)
#projection matrix
w = np.hstack((eigen_pair[0][1][:, np.newaxis], eigen_pair[1][1][:, np.newaxis]))
display(w)

new_X = X_scaled.dot(w)   # 142 by 13 and 13 by 2
display(new_X)
display(new_X.shape)
   
for l in np.unique(y):
    plt.scatter(new_X[y==l, 0], new_X[y==l, 1])
plt.title("Transformed Features of Dim=2")
plt.xlabel('PCA1')
plt.ylabel("PCA2")
plt.show()
```

# Locally Linear Embedding

```python
from sklearn.manifold import LocallyLinearEmbedding
```

```python
lle=LocallyLinearEmbedding(n_neighbors=10,n_components=2)
```

## Transforming the data

```python
X_lle=lle.fit_transform(X)
display(X_lle)
```

## Visualisation

```python
plt.scatter(X_lle[:,0],X_lle[:,1],c=y);
```

```python
# Testing Neighbourhoods from 5-15

for i in range(5,16):
    lle=LocallyLinearEmbedding(n_neighbors=i,n_components=2)
    X_lle=lle.fit_transform(X)
    plt.scatter(X_lle[:,0],X_lle[:,1],c=y)
    plt.title(f"Neighbors = {i}")
    plt.show()
```

## The model with 8 neighbors is best able to seperate the data points in the 3 classes. This helps us better visualize the data. The transformed data can be used for clustering and visualizing the output of the clustering algorithm thus improving performance. 

# TSNE

```python
from sklearn.manifold import TSNE
import seaborn as sns
tsne = TSNE(n_components = 2, perplexity = 30)
```

```python
X_tsne = tsne.fit_transform(X)
display(X_tsne.shape)
```

```python
display(X_tsne[0])
display(X_tsne[1])
```

## Visualisation of the transformed points

```python
plt.scatter(X_tsne[:, 0], X_tsne[:, 1]);
```

```python
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c = y.astype(float));
```

## Creating a DF for better visualisation

```python
X_df = pd.DataFrame({"X0": X_tsne[:, 0],
                    "X1": X_tsne[:, 1],
                    "Label": y})
display(X_df)
```

```python
sns.lmplot(data = X_df, x = "X0", y = "X1", hue = "Label");
```

# Trying our Different values of Perplexity from 10-200

```python
for i in range(20,201,20):
    tsne = TSNE(n_components = 2, perplexity = i)
    X_tsne = tsne.fit_transform(X)
    plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c = y.astype(float))
    plt.title(f"Perplexity = {i}")
    plt.show()
```

## Higher perplexity seems to help the model segregate the labels very slightly upto perlexity of 100. After which the results don't change much.
## The model is not able to segregate the data into clean clusters, the data is scattered in the 2 dimensional space. The performance is worse than LLE.

## The LLE model is best able to segregate the data in the lower dimensional space. This is seem clearly by the plots in the 2D space which show that 
1) PCA analysis does not segregate the data at all
2) TSNE in the lower dim space has the points all scattered throughout the plot with no clear clusters.
3) Only LLE is able to create clusters in the lower dim space, even the though they are somewhat overlapping and the performance is less than ideal.


## PCA is the least effective in segregating the data into clusters. This is because PCA is not a segregating technique, it only reduces the dimensions of the feature space based on the total variance explained by the PCA components, which can be seen in the plots of PCA where the variance along PCA1 is highest and PCA2 is second highest. Thus PCA is not trying to segregate the data, hence it's performance in segregating is the worst among PCA, LLE and TSNE

## Some ways to improve performance of dimensionality reduction are

1) Feature selection: Instead of using all the features available, we can select a subset of the most important features based on their relevance to the problem at hand. This can improve performance by reducing noise.

2) Other algorithms: We can try other algorithms for dim reductions to see if any of them performan better.

3) Parameter tuning: Adjusting the hyperparameters of the algorithm can also improve efficiency, like adjusting number of neighbors in LLE or Perplexity in TSNE.
