```python
nlp = spacy.load('en_core_web_sm')
# nlp.add_pipe('sentencizer')
doc = nlp(text)
type(doc)
```

```python
#count number of sentences
sentence_count = 0
for sent in doc.sents:
    sentence_count+=1
print(f"Sentence Count = {sentence_count}")
```

```python
#count number of tokens
token_count = 0
for sent in doc.sents:
    for token in sent:
        token_count+=1
print(f"Token Count = {token_count}")
```

```python
#Parts Of Speech for each token
for sent in doc.sents:
    for token in sent:
        print(token, '==>', token.pos_)
```

```python
#tokens and its POS, where the token (i) not a stop word (ii) not a punctuation and (iii)  not a number
for sent in doc.sents:
    for token in sent:
        if((not token.is_stop) and (not token.is_punct) and (not token.like_num)):
            print(token, '==>', token.pos_)
```

```python
#Count number of nouns and verbs in the text
nouns = 0
verbs = 0
for sent in doc.sents:
    for token in sent:
        if(token.pos_ == 'NOUN'):
            nouns+=1
        if(token.pos_ == 'VERB'):
            verbs+=1
print(f"Nouns = {nouns}")
print(f"Verbs = {verbs}")
```

```python
#Display Named Entities and Their Respective Types
for ent in doc.ents:
    print(ent.text,'==>',ent.label_)
```

```python
# Display Entities and their types
from spacy import displacy

displacy.render(doc, style='ent')
```

```python
#Create a DataFrame consisting of entities, types and explanation of types
import pandas as pd

cols=['Entities','Type', 'Explain_Type']
rows=[]
for entity in doc.ents:
    row=entity,entity.label_,spacy.explain(entity.label_)
    rows.append(row)

entity_df=pd.DataFrame(rows,columns=cols)
entity_df
```

```python
#Plot counts of each type
import matplotlib.pyplot as plt
ent_list=[]
for ent in doc.ents:
    ent_list.append(ent.label_)
print(ent_list)
from collections import Counter
counter = Counter(ent_list)
print(counter)
plt.bar(counter.keys(), counter.values())
plt.xticks(rotation=90)
plt.show()
```

# Person Type Appears most Often in the Document

```python
#Plot counts of each person
import matplotlib.pyplot as plt
person_list=[]
for ent in doc.ents:
    if(ent.label_=="PERSON"):
        person_list.append(str(ent))
print(person_list)
counter = Counter(person_list)
print(counter)
plt.bar(counter.keys(), counter.values())
plt.xticks(rotation=90)
plt.show()
```

```python
#Matching words
target_words = {'depict', 'reveal', 'discover', 'direct', 'narrate', 'illustrate', 'portray'}
target_occurrences = []

for token in doc:
    if token.lemma_ in target_words:
        target_occurrences.append(token)

word_list = []
for token in target_occurrences:
    word_list.append(token.lemma_)
count = Counter(word_list)
print(count)
```
